```python
import numpy as np

import os
import re
from PIL import Image
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
f = open('clean_adjectives.txt', 'r', encoding='utf8')
txt = f.readlines()
f.close()
```

```python
new_txt = []
for i in txt:
    new_txt.append(i.replace('\n', ''))
```

```python
from collections import Counter
Counter(new_txt).most_common()
```

```python
mask = np.array(Image.open("picture.png"))
```

```python
#stopword 리스트 작성 
stop = ['없다', '아니다', '같다', '이다', '있다', '이렇다' ]
stopwords = set(stop)
```

```python
str_txt = ''
for i in new_txt:
    str_txt = str_txt + i + ' ' #공백 띄어쓰기 해줘야 함 
```

```python
wc = WordCloud(background_color='white', font_path ='‪C:/Windows/Fonts/NanumGothicExtraBold.ttf',max_words=1000, mask=mask, stopwords=stopwords, margin=10,
               random_state=1,contour_color='black', colormap='summer').generate(str_txt)

wc.to_file("a_new_hope.png")
plt.axis("off")
plt.figure(figsize=(10,10))
plt.title("Default colors")
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
