```python
##명사 추출 
from konlpy.tag import Okt
import numpy as np
okt = Okt()


clean_nouns = []

for k in words_list:
  nouns = okt.nouns(k)

  for noun in nouns:
    clean_nouns.append(noun)

print(clean_nouns)
len(clean_nouns)
```

```python
# 만들어진 리스트 텍스트 파일로 저장 
for i in clean_nouns:
  f = open('clean_nouns.txt','a', encoding= 'utf8')
  f.write(i+ '\n')
  f.close()
```
